**Instructions**
Your objective is to delve into the World Happiness Report dataset, which includes columns like ‘Country’, ‘Year’, ‘Happiness_Score’, ‘GDP_per_Capita’, ‘Social_Support’, ‘Healthy_Life_Expectancy’, ‘Freedom_to_Make_Life_Choices’, ‘Generosity’, and ‘Perceptions_of_Corruption’.  

**Load and Inspect Data:**  
Utilize pandas to load the World Happiness Report Dataset. (https://www.kaggle.com/datasets/unsdsn/world-happiness)  
Examine the data to grasp its structure and contents of the year 2019.  

**Data Preprocessing:**  
Address missing values and adjust data types as necessary.  

**Social Support and Happiness:**  
Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.  

**Regional Comparisons of GDP and Health:**  
Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.  
Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.

**Loadind and inspection**

```python
import pandas as pd
df = pd.read_csv('2019.csv')
df.head()
```

**Data Preprocessing**

```python
blank_val = df.isnull().values.any()
print('Presence of blank rows:', blank_val)
print(df.dtypes)
```

**Scatter plot of 'Social support' against 'Score'**

```python
import matplotlib.pyplot as plt
plt.scatter(df['Social support'], df['Score'])
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.title('Social Support vs. Happiness Score')
plt.show
```

**GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions (Subplot)**

```python
grouped = df.groupby("Country or region")[["GDP per capita", "Healthy life expectancy"]].mean()
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].bar(grouped.index, grouped["GDP per capita"])
axes[0].set_title("GDP per capita by region")
axes[0].set_xlabel("Country or region")
axes[0].set_ylabel("GDP per capita")
axes[0].tick_params(axis='x', rotation=90)
axes[1].bar(grouped.index, grouped["Healthy life expectancy"])
axes[1].set_title("Healthy life expectancy by region")
axes[1].set_xlabel("Country or region")
axes[1].set_ylabel("Life expectancy")
axes[1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

**Tweaks for better readability of plots using flexible figsize**

```python
n = len(df["Country or region"].unique())
fig, axes = plt.subplots(1, 2, figsize=(max(10, n * 0.4), 6))
grouped = df.groupby("Country or region")[["GDP per capita", "Healthy life expectancy"]].mean()
axes[0].bar(grouped.index, grouped["GDP per capita"])
axes[0].set_title("GDP per capita by region")
axes[0].set_xlabel("Country or region")
axes[0].set_ylabel("GDP per capita")
axes[0].tick_params(axis='x', rotation=90)
axes[1].bar(grouped.index, grouped["Healthy life expectancy"])
axes[1].set_title("Healthy life expectancy by region")
axes[1].set_xlabel("Country or region")
axes[1].set_ylabel("Life expectancy")
axes[1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

**Observation on how economic strength relates to health outcomes in different regions**

```python
n = len(grouped.index)
fig, ax1 = plt.subplots(figsize=(max(10, n * 0.6), 6))
ax1.bar(grouped.index, grouped["GDP per capita"], color="blue", alpha=0.7)
ax1.set_ylabel("GDP per capita", color="blue")
ax1.set_xlabel("Region")
ax1.tick_params(axis='x', rotation=90)
ax2 = ax1.twinx()
ax2.plot(grouped.index, grouped["Healthy life expectancy"], color="red", marker="o")
ax2.set_ylabel("Healthy life expectancy", color="red")
plt.title("GDP per capita vs Healthy life expectancy by Region")
plt.tight_layout()
plt.show()
```

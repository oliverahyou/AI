```python
# Step 1: Transforming the String into a 2D List
MATRIX_STR = '''
7ir
Tsi
h%x
i ?
sM#
$a
#t%'''
matrix = MATRIX_STR.split()
print(matrix)
```

```python
# Step 2: Iterate through columns
max_len = max(len(row) for row in matrix)
for col in range(max_len):
    for row in matrix:
        if col < len(row):
          char = row[col]
          print(char, end='')
```

```python
# Step 3: Filter alpha characters
matrix = ['7ir', 'Tsi', 'h%x', 'i', '?', 'sM#', '$a', '#t%']
max_len = max(len(row) for row in matrix)
result = ""
for col in range(max_len):
    for row in matrix:
        if col < len(row):
            char = row[col]
            if char.isalpha():
                result += char
print(result)
```

```python
# Step 4: Replace symbols with spaces
matrix = ['7ir', 'Tsi', 'h%x', 'i', '?', 'sM#', '$a', '#t%']
max_len = max(len(row) for row in matrix)
result = ""
for col in range(max_len):
    for row in matrix:
        if col < len(row):
            char = row[col]
            if char.isalpha():
                result += char
            else:
                result += ' '
print(result)
```
